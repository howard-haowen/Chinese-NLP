```python
import pandas as pd
```

```python
wordlists = pd.read_excel("input.xlsx", 
                   sheet_name=None,
                   engine='openpyxl')
len(wordlists)
```

```python
wordlists.keys()
```

## 匯總

```python
levels = list(wordlists.keys())[:-1]
levels
```

```python
wordlists['準備級一級']
```

```python
dfs = [wordlists[key] for key in levels]
data = pd.concat(dfs)
data
```

## 過濾

```python
cols = ["展開表", "漢語拼音", "注音", "任務領域", "詞條編號"]
data = data[cols]
data
```

## 補缺

```python
data = data.fillna(method="ffill") # forward fill
data
```

## 去重

```python
query = '下'
data.query('展開表 == @query')
```

```python
key_col = '展開表'
data.drop_duplicates(subset=key_col, 
                     keep='first', 
                     inplace=True, 
                     ignore_index=True)
data
```

```python
data.rename(columns={"展開表":"詞彙"}, inplace=True)
data
```

```python
data.reset_index(drop=True, inplace=True)
data
```

## 變形

```python
word_levels = data['詞條編號'].apply(lambda x: x.split("-")[0])
word_levels
```

```python
mapping = {
    "準1": "準備1",
    "準2": "準備2",
    "入": "入門",
    "基": "基礎",
    "進": "進階",
    "高": "高階",
    "流": "流利",
}
```

```python
renamed_levels = word_levels.map(mapping)
renamed_levels
```

```python
data['詞條分級'] = renamed_levels
data
```

## 繪圖

```python
import plotly.express as px
```

```python
data['任務領域'].value_counts()
```

```python
domain = data['任務領域'].value_counts()
fig = px.pie(data, 
             values=domain.values, 
             names=domain.index, 
             title='任務領域圓餅圖')
fig.show()
```

```python
level = data['詞條分級'].value_counts()
fig = px.pie(data, 
             values=level.values, 
             names=level.index, 
             title='詞條分級圓餅圖')
fig.show()
```

## 儲存
